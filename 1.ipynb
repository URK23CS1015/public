```python
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def __add__(self, other):
        
        return Rectangle(self.length + other.length, self.breadth + other.breadth)

    def __str__(self):
        return f"Length is {self.length} and Breadth is {self.breadth}"

    def __eq__(self, other):
        return self.length == other.length and self.breadth == other.breadth

    def __lt__(self, other):
        return self.length * self.breadth < other.length * other.breadth

    def __gt__(self, other):
        return self.length * self.breadth > other.length * other.breadth

    def __le__(self, other):
        return self.length * self.breadth <= other.length * other.breadth

    def __ge__(self, other):
        return self.length * self.breadth >= other.length * other.breadth


r1 = Rectangle(10, 5)
r2 = Rectangle(20, 6)


r3 = r1 + r2

print(r3)

print(f"r1 == r2: {r1 == r2}")
print(f"r1 < r2: {r1 < r2}")
print(f"r1 > r2: {r1 > r2}")
print(f"r1 <= r2: {r1 <= r2}")
print(f"r1 >= r2: {r1 >= r2}")
```

```python
class Employee:
    def __init__(self, empid, name, department, designation, experience, basic_pay):
        self.empid = empid
        self.name = name
        self.department = department
        self.designation = designation
        self.experience = experience
        self.basic_pay = basic_pay
        self.da = 0.10 * basic_pay
        self.hra = 0.05 * basic_pay
        self.epf = 0.05 * basic_pay
        self.tax = 0.10 * basic_pay

    def calculate_net_salary(self):
        return self.basic_pay + self.da + self.hra - self.epf - self.tax

    def __str__(self):
        return f"Employee ID: {self.empid}\nName: {self.name}\nDepartment: {self.department}\nDesignation: {self.designation}\nExperience: {self.experience} years\nBasic Pay: {self.basic_pay}\nNet Salary: {self.calculate_net_salary()}"

def get_employee_details():
    empid = input("Enter Employee ID: ")
    name = input("Enter Employee Name: ")
    department = input("Enter Department: ")
    designation = input("Enter Designation: ")
    experience = float(input("Enter Experience (in years): "))
    basic_pay = float(input("Enter Basic Pay: "))
    return Employee(empid, name, department, designation, experience, basic_pay)

def display_employee_details(employee_list):
    for emp in employee_list:
        print(emp)
        print("-" * 40)

def main():
    employee_list = []

    while True:
        print("\nEmployee Payroll Management System")
        print("1. Add Employee")
        print("2. Edit Employee Details")
        print("3. Delete Employee")
        print("4. Display Employee Details")
        print("5. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            employee = get_employee_details()
            employee_list.append(employee)
            print("Employee added successfully!")

        elif choice == 2:
            empid = input("Enter Employee ID to edit details: ")
            for emp in employee_list:
                if emp.empid == empid:
                    print("Current Employee Details:")
                    print(emp)
                    print("\nEnter new details:")
                    new_employee = get_employee_details()
                    employee_list.remove(emp)
                    employee_list.append(new_employee)
                    print("Employee details updated successfully!")
                    break
            else:
                print("Employee not found!")

        elif choice == 3:
            empid = input("Enter Employee ID to delete: ")
            for emp in employee_list:
                if emp.empid == empid:
                    employee_list.remove(emp)
                    print("Employee deleted successfully!")
                    break
            else:
                print("Employee not found!")

        elif choice == 4:
            display_employee_details(employee_list)

        elif choice == 5:
            print("Exiting the program. Thank you!")
            break

        else:
            print("Invalid choice! Please select a valid option.")

if __name__ == "__main__":
    main()
```
